### Data Encoding
Data Encoding refers to the process of converting categorical data into meaningfull numerical information

```python
import pandas as pd
```

```python
df = pd.DataFrame({
    'color' : ['red','blue', 'red','green','green', 'blue']     
})
df
```

##### OneHotEncoding or Ordinal encoding technique
Is in which each category is presented as 0 or 1 if the value belons to a particular category it is stored as 1 otherwise 0 <br>
Cons: We never use it for too many categories

```python
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder()
encoded = encoder.fit_transform(df[["color"]])
encoded_df = pd.DataFrame(encoded.toarray(),columns=encoder.get_feature_names_out(['color']))
```
