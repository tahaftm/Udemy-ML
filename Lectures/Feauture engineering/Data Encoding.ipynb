### Data Encoding
Data Encoding refers to the process of converting categorical data into meaningfull numerical information

```python
import pandas as pd
```

```python
df = pd.DataFrame({
    'color' : ['red','blue', 'red','green','green', 'blue']     
})
df
```

##### OneHotEncoding or Ordinal encoding technique
Is in which each category is presented as 0 or 1 if the value belons to a particular category it is stored as 1 otherwise 0 <br>
Cons: We never use it for too many categories

```python
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder()
encoded = encoder.fit_transform(df[["color"]])
encoded_df = pd.DataFrame(encoded.toarray(),columns=encoder.get_feature_names_out(['color']))
```

```python
encoded_df
```

##### Label Encoding
Is in which each category has a particular number assigned to it.

```python
from sklearn.preprocessing import LabelEncoder
lbl_encoder = LabelEncoder()
encoded = lbl_encoder.fit_transform(df[['color']])
```

```python
lbl_encoded_df = pd.DataFrame(encoded, columns= ['color'])
```

```python
lbl_encoded_df
```

##### Ordinal Encoding
Is used when we are in need of assigning some ranks such as:<br>
High School: 4<br>
Matriculation: 3<br>
Secondary: 2<br>
Primary: 1<br>

```python
from sklearn.preprocessing import OrdinalEncoder
sizes = pd.DataFrame({"size" : ["small", "large", "large", "medium", "small", "medium"]})
ord_encoder = OrdinalEncoder()
ord_encoder.fit_transform(sizes[['size']])
```
