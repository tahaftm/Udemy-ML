### Handling imbalance dataset
Up sampling<br>
Down sampling

```python
import pandas as pd
import numpy as np
```

```python
np.random.seed(123)
```

```python
n_samples = 1000
class_0_ratio = 0.9
n_class_0 = int(n_samples * class_0_ratio)
n_class_1 = n_samples - n_class_0
n_class_0,n_class_1
```

```python
class_0 = pd.DataFrame({
    'feature_1' : np.random.normal(loc = 0, scale = 1, size=n_class_0),
    'feature_2' : np.random.normal(loc = 0, scale = 1, size=n_class_0), 
    'target' : [0] * n_class_0
    })
class_1 = pd.DataFrame({
    'feature_1' : np.random.normal(loc = 0, scale = 1, size=n_class_1),
    'feature_2' : np.random.normal(loc = 0, scale = 1, size=n_class_1), 
    'target' : [1] * n_class_1
    })
```

```python
df = pd.concat([class_0,class_1]).reset_index(drop=True)
df.tail()
```

```python
df['target'].value_counts()
```

```python
minority_df = df[df['target'] == 1]
majority_df = df[df['target'] == 0]
```

## Upsampled

```python
from sklearn.utils import resample
minority_df_resampled = resample(minority_df, replace=True, n_samples=len(majority_df), random_state=42)
minority_df_resampled.shape
```

```python
df_resampled = pd.concat([majority_df, minority_df_resampled])
df_resampled['target'].value_counts()
```

## Downsampled

```python
class_0 = pd.DataFrame({
    'feature_1' : np.random.normal(loc = 0, scale = 1, size=n_class_0),
    'feature_2' : np.random.normal(loc = 0, scale = 1, size=n_class_0), 
    'target' : [0] * n_class_0
    })
class_1 = pd.DataFrame({
    'feature_1' : np.random.normal(loc = 0, scale = 1, size=n_class_1),
    'feature_2' : np.random.normal(loc = 0, scale = 1, size=n_class_1), 
    'target' : [1] * n_class_1
    })
```

```python
df = pd.concat([class_0,class_1]).reset_index(drop=True)
```

```python
df['target'].value_counts()
```

```python
minority_df = df[df['target'] == 1]
majority_df = df[df['target'] == 0]
```

```python
from sklearn.utils import resample
majority_df_resampled = resample(majority_df, replace=False, n_samples=len(minority_df), random_state=42)
majority_df_resampled.shape
```

```python
df_resampled = pd.concat([minority_df,majority_df_resampled])
df_resampled.target.value_counts()
```
