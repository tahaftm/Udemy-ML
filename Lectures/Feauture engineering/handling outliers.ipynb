### Handling Outliers

##### Calculating 5 number summary

```python
import numpy as np
```

```python
marks = [-100, 89,78,67,88,98,77,66,76,56,45,76,87, 190]
minimum, Q1, median, Q3, maximum=np.quantile(marks, [0, 0.25, 0.5, 0.75, 1.0])
```

```python
IQR = Q3 - Q1
```

```python
lower_fence = int(IQR - Q1*1.5)
higher_fence = int(IQR + Q3*1.5)
```

```python
lower_fence, higher_fence
```

```python
print("Minimum: ", minimum)
print("Q1: ", Q1)
print("Median: ", median)
print("Q3: ", Q3)
print("Maximum: ", maximum)
```

```python
import seaborn as sns
sns.boxplot(marks)
```
