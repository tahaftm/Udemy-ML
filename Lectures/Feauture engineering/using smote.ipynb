### Handling imbalance data using smote

In normal upsampling technique the problem us that we increase data points w.r.t the previous data points we just increase the overall number wereas the actual data points remain the same 

In smote technique the two data points are joined together and that line joing both ends have more data points lying on that line/

```python
import pandas as pd
from sklearn.datasets import make_classification
```

```python
X,y = make_classification(n_samples=1000, n_features= 2, n_redundant=0, n_clusters_per_class=1, weights = [0.90],random_state=12)
```

```python
df1 = pd.DataFrame(X, columns=['feature_1', 'feature_2'])
df2 = pd.DataFrame(y,columns = ['target'])
final_df = pd.concat([df1,df2],axis=1)
final_df.target.value_counts()
```

```python
import matplotlib.pyplot as plt
plt.scatter(final_df['feature_1'],final_df['feature_2'], c = final_df['target'])
```

```python
from imblearn.over_sampling import SMOTE
over_sampling = SMOTE()
x,y = over_sampling.fit_resample(final_df[['feature_1', 'feature_2']], final_df['target'])
x
```

```python
df1=pd.DataFrame(x)
df2=pd.DataFrame(y)
oversample_df=pd.concat([df1,df2],axis=1)
```

```python
plt.scatter(oversample_df['feature_1'], oversample_df['feature_2'], c = oversample_df['target'])
```
