### Multiprocessing

Multiprocessing refers to the phenomenon of running different processes `an instance of the program` at the same time using different `cores` of the computer

```python
import multiprocessing
import time

def print_numbers():
    for i in range(1, 6):
        time.sleep(1)
        print(f"Number: {i}", flush=True)

def print_letters():
    for i in 'abcde':
        time.sleep(1.5)
        print(f"Letter: {i}", flush=True)

if __name__ == "__main__":
    time_start = time.time()

    t1 = multiprocessing.Process(target=print_numbers)
    t2 = multiprocessing.Process(target=print_letters)

    t1.start()
    t2.start()

    t1.join()
    t2.join()

    time_ended = time.time()
    print(f"The time taken to complete the operation was: {time_ended - time_start}s")
```
