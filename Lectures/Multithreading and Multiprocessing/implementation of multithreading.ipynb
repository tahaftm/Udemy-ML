### Multithreading

##### When to use multithreading?
-> I/O bound task: Task that requires more input output operations.<br>
-> Concurent Execution: Multiple executions at once.

```python
import time
```

### Before using threads:

```python
def print_numbers():
    for i in range(1,6):
        time.sleep(2)
        print(f"Number: {i}")
    
def print_letters():
    for i in 'abcde':
        time.sleep(2)
        print(f"Letter: {i}")

time_start = time.time()
print_numbers()
print_letters()
time_ended = time.time()
taken_time = time_ended - time_start
print(f"The time taken to complete the operation was: {taken_time}s")
```

### After using Threads:


```python
import threading

def print_numbers():
    for i in range(1,6):
        time.sleep(2)
        print(f"Number: {i}")
    
def print_letters():
    for i in 'abcde':
        time.sleep(2)
        print(f"Letter: {i}")

time_start = time.time()

## creating threads:
t1 = threading.Thread(target=print_numbers)
t2 = threading.Thread(target=print_letters)

## staring the threads:
t1.start()
t2.start()

## waiting for the thread to complete
t1.join()
t2.join()

time_ended = time.time()
taken_time = time_ended - time_start
print(f"The time taken to complete the operation was: {taken_time}s")
```
