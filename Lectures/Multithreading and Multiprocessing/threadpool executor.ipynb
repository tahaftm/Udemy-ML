### ThreadPoolExecuter
<br>
We use thread pool executor because they are easier to manage when creating multiple threads and keeping track of threads. They make the process of making,managing and tracking threads more efficiently.

```python
from concurrent.futures import ThreadPoolExecutor
import time
```

```python
def compute_squares(i):
    time.sleep(2)
    return f"The square of {i}: {i*i}"

numbers = [1,3,45,6,8,9,54,2,2]
if __name__ == '__main__':
    with ThreadPoolExecutor(max_workers=3) as executor:
        result = executor.map(compute_squares,numbers)
    
    for i in result:
        print(i)
```
