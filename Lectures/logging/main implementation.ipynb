### Logging:
`Logging` is a method of displaying information with different severity levels. We first have to import it and then configure it with `basicConfig` function and then we can write the messages with severity levels

```python
import logging
logging.basicConfig(
    level=logging.DEBUG,
    format='%(asctime)s-%(name)s-%(levelname)s-%(message)s',
    datefmt= '%d-%m-%Y %H-%M-%S',
    force = True
)

logging.info("This is a info message")
logging.debug("This is a debug message")
logging.error("This is a error message")
logging.critical("This is a critical message")
logging.warning("This is a warning message")
```

```python
import logging
import time
logging.basicConfig(
    filename="logs.log",
    filemode="w",
    level=logging.DEBUG,
    format='%(asctime)s-%(name)s-%(levelname)s-%(message)s',
    datefmt= '%d-%m-%Y %H-%M-%S',
    force = True
)

logging.info("This is a info message")
logging.debug("This is a debug message")
logging.error("This is a error message")
time.sleep(5)
logging.critical("This is a critical message")
logging.warning("This is a warning message")
```
