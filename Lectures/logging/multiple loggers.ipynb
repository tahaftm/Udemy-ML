### Multiple Loggers:

```python
import logging
```

```python
logger1 = logging.getLogger("Module 1")
logger1.setLevel(logging.ERROR)

logger2 = logging.getLogger("Module 2")
logger2.setLevel(logging.DEBUG)
```

```python
logging.basicConfig(
    level=logging.DEBUG,
    format='%(asctime)s-%(name)s-%(levelname)s-%(message)s',
    datefmt= '%d-%m-%Y %H-%M-%S',
    force = True
)

logger1.warning("This is a info message")
logger2.debug("This is a debug message")
```
