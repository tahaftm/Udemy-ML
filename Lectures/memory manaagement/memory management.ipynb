### Memory Management

### Reference method:

A method of managing memory <br>
Each object refers to a memory location. If the reference count drops to 0 the memory for that object becomes zero

A function is present in sys module from which we can get the current number of references for any object

```python
import sys
a = []
print(sys.getrefcount(a))
```

one reference for the list object and one for the function itself


```python
b = a
print(sys.getrefcount(b))
```

```python
del b
print(sys.getrefcount(a))
```

##### Garbage collector:
It is used to handle reference cycles

`Reference cycles:`  When objects refer each other and preventing reference count from reaching 0 <br>
`How?: ` if 'a' points to 'b' then b has its own reference count along with 'a' and vice versa so deleting one of them do not deletes the other one  

```python
import gc
gc.enable()
```

```python
gc.disable()
```

```python
gc.collect()
```

##### Memory management practices
1. Using Local Variables: Local variable have shorteer lifespan than global variables<br>
2. Avioding Circular references: It means to variables refer each other<br>
3. Using Generators: Generators are used as iterator giving one value at a time<br>
4. Deleting the variables: Use the `del` keyword to delete the objects<br>
5. Profile memory usage: Using memory profiling tools like tracemalloc and memory_profiler to identify memory leak and optimize memory usage

```python
import sys
class MyObject():
    def __init__(self, name):
        self.name = name
        print(F"{self.name} was created")

    def __del__(self):
        print(F"{self.name} was deleted")

obj1 = MyObject("obj1")
obj2 = MyObject("obj2")

obj1.ref = obj2
obj2.ref = obj1

del obj1
del obj2

gc.collect()
```
