### Assignment # 1


```python
class Car:
    def __init__(self,make,model,year):
        self.make = make
        self.model = model
        self.year = year
    def display_car(self):
        print(f"This is a {self.make} car of model {self.model} from year {self.year}")
```

```python
audi = Car("Audi","A4",2020)
print(audi.make)
print(audi.model)
print(audi.year)
```

```python
audi.display_car()
```

```python
class Car:
    def __init__(self,make,model,year):
        self.make = make
        self.model = model
        self.year = year
    def display_car(self):
        print(f"This is a {self.make} car of model {self.model} from year {self.year}")
    def engine_start(self):
        print(f"{self.model} has started its engine")
```

```python
bmw = Car("BMW","X5",2021)
bmw.engine_start()
```
