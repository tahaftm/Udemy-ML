### Assignment # 6

```python
def testing_exceptions(input_):
    try:
        num = int(input_)
        try: 
            quotient = 4/num
            print(quotient)
        except ZeroDivisionError:
            print("The number should be divided by non zero.")
        finally:
            print("Done division!")
    except ValueError:
        print("Please enter a number.")
    finally:
        print("Done Execution!")
```

```python
testing_exceptions(4)
```

```python
testing_exceptions('0')
```

### Assignment # 9

```python
import json
def json_parser(string):
    try:
        json_object = json.loads(string)
        with open("exceptionhandling.json","w") as file:
            json.dump(json_object,file)
    except (json.JSONDecodeError,AttributeError):
        print("Invalid JSON string.")
```

```python
json_parser("{'name': 'Taha'}")
```

```python
json_parser('{"price": 99.5, "available": false}')
```

```python
json_parser('{"name": "Taha", "age": 22}')
```

### Assignment # 10

```python
class NegativeNumberError(Exception):
    pass
def negative_exceptions(lists):
    for i in lists:
        try:
            if i < 0:
                raise NegativeNumberError
        except NegativeNumberError:
            print("Negative number found:", i)    
```

```python
negative_exceptions([4, -3, 2, -1, 0, 5, -6])
```

### Assignment # 12

```python
class Calcualations:
    def __init__(self,num,den):
        self.num = num
        self.den = den
    def divide(self):
        try:
            result = self.num / self.den
            print("The result is:", result)
        except ZeroDivisionError:
            print("Error: Division by zero is not allowed.")

calc1 = Calcualations(10,2)
calc1.divide()
```

```python
calc2 = Calcualations(10,0)
calc2.divide()
```
